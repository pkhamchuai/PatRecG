```python
import matplotlib.pyplot as plt
import numpy as np

# specify a larger font size for plots
plt.rcParams['font.size'] = 16

```

```python
# define a density function
def f(x, h):
    if x >= 0 and x <= 2/h:
        return h - (h*h*x/2)
    else:
        return 0
    
```

```python
# create a linspace
x = np.linspace(-2, 5, 100)

# plot 5 differrent graphs for different h values
for h in [0.5, 1, 2, 3, 4]:
    y = [f(i, h) for i in x]
    plt.plot(x, y, label='h = ' + str(h))
    # show the legend
    plt.legend()
    # show grid lines
    plt.grid(True)
    # plot labels, larger text
    plt.title('f(x|h) vs x')
    plt.xlabel('x')
    plt.ylabel('f(x|h)')
```

```python
# plot the graphs of f versus 5 different h values for 5 fixed x values
x_ = [-1, 0, 1, 2, 3]
h = np.linspace(0, 3, 100)

for x in x_:
    y = [f(x, i) for i in h]
    plt.plot(h, y, label='x = ' + str(x))
    # show the legend
    plt.legend()
    # show grid lines
    plt.grid(True)
    # show labels for x and y axis
    plt.title('f(x|h) vs h')
    plt.xlabel('h')
    plt.ylabel('f(x|h)')
```
